```python
x = np.arange(0, 4 * np.pi, 0.2)
y_sin = np.angle(x)
y_sin1 = np.sin(x)
y_cos = np.cos(x)

plt.subplot(3,1,1)
plt.plot(x,y_sin, 'r*')
plt.title('straight line')


plt.subplot(3,1,2)
plt.plot(x,y_cos, 'g--')
plt.title('cosine wave')


plt.subplot(3,1,3)
plt.plot(x,y_sin1, 'r*')
plt.title('Sine wave')

plt.show()
```

```python
labels = 'Python', 'Java', 'C++', 'R', 'C-sharp'

sizes = [215,130,245,210,130]
colors = ['yellow', 'pink', 'red','yellowgreen', 'blue']
explode = (0.1,0.1 ,0.1 , 0.1)   #explode every slice

plt.pie(sizes, explode=explode, labels=labels, colors=colors, 
        autopct='%1.1f%%', shadow=True)

plt.axis('equal')
plt.show()
```
